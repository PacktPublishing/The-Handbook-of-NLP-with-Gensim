```python
import numpy as np
import scipy.stats as stats
%matplotlib inline
import matplotlib.pyplot as plt

binomial=stats.binom(n=10,p=0.2)
binomial.rvs(size=1,random_state=123)
```

```python
binomial.rvs(size=10,random_state=123)
```

```python
binomial.pmf([1,2,3,4,5,6,7,8,9])
```

```python
t=np.arange(0,10)
binomial_sample=binomial.rvs(size=1000,random_state=123)
plt.hist(binomial_sample,bins=10,range=(0,10),density=True,align='left',rwidth=0.75,label="P=0.2",alpha=0.75)
plt.title("Binomial Distribution")
plt.ylabel("Density")
plt.legend()
plt.xticks(t)
plt.xlabel("Number of successes")
plt.show()
```

### (B) Multinomial Distribution

```python
import numpy as np
import scipy.stats as stats
multinomial=stats.multinomial(n=10,p=[1/12, 1/3, 1/12, 1/6, 1/6, 1/6])
multinomial.rvs(size=10,random_state=143)
```

```python
multinomial_sample=multinomial.rvs(size=1000,random_state=143)
multinomial_sample
```

```python
t=np.arange(0,9)
plt.hist(multinomial_sample,bins=10,range=(0,9),density=True,align='left',
         rwidth=0.9,label=['p1=1/12','p2=1/3','p3=1/12','p4=1/6','p5=1/6','p6=1/6'],alpha=0.75)
plt.title("Multinomial Distribution")
plt.ylabel("Density")
plt.legend()
plt.xticks(t)
plt.xlabel("Number of successes")
plt.show()
```

### (C) Beta distribution

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.beta.html

```python
import numpy as np
import scipy.stats as stats
%matplotlib inline
import matplotlib.pyplot as plt
a, b = 0.5, 0.5
my_beta = stats.beta(a, b)
my_beta
```

```python
my_beta.rvs(size=10,random_state=143)
```

```python
beta_sample=my_beta.rvs(size=1000,random_state=143)
beta_sample
```

```python
t=np.arange(0,2)
plt.hist(beta_sample,bins=100,range=(0,1),density=True,align='left',
         rwidth=1,label=['a = 0.5, b = 0.5'],alpha=0.75)
plt.title("Beta Distribution")
plt.ylabel("Density")
plt.legend()
plt.xticks(t)
plt.xlabel("Probability")
plt.show()
```

### (C) Dirichlet Distribution

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
alpha = [1,1,1]
D = stats.dirichlet(alpha).rvs(500)
D
```

```python
def plot_Dirichlet(alpha):
    D = stats.dirichlet(alpha).rvs(500)
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(projection='3d')
    plt.title(r'$\alpha$ = {}'.format(alpha), fontsize=18)
    ax.scatter(D[:, 0], D[:, 1], D[:, 2],c='b')
    ax.view_init(azim=30)
    ax.set_xlabel(r'$\theta_1$', fontsize=14)
    ax.set_ylabel(r'$\theta_2$', fontsize=14)
    ax.set_zlabel(r'$\theta_3$', fontsize=14)
    plt.tight_layout()
    plt.show()

plot_Dirichlet([1,1,1])
```

```python
plot_Dirichlet([1,1,5])
```

```python
plot_Dirichlet([0.1,0.1,0.1])
```

```python
plot_Dirichlet([5,5,5])
```
